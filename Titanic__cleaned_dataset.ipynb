```python
# STEP 1: Import required libraries
import pandas as pd
import numpy as np

# STEP 2: Upload file in Google Colab
from google.colab import files
uploaded = files.upload()

# STEP 3: Load the dataset
# Replace 'titanic.csv' with your actual file name if different
df = pd.read_csv('Titanic-Dataset.csv')

# STEP 4: View first 5 rows
print("First 5 rows of dataset:")
display(df.head())

# STEP 5: Check dataset information
print("\nDataset Info:")
df.info()

# STEP 6: Check missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# ===============================
# DATA CLEANING STARTS HERE
# ===============================

# STEP 7: Handle missing values in Age column
# Fill missing Age values with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# STEP 8: Handle missing values in Embarked column
# Fill with most frequent value (mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# STEP 9: Handle Cabin column
# Too many missing values, so drop the column
df.drop(columns=['Cabin'], inplace=True)

# STEP 10: Drop columns not useful for ML
df.drop(columns=['PassengerId', 'Name', 'Ticket'], inplace=True)

# STEP 11: Convert categorical columns into numeric
# Sex column (male=0, female=1)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Embarked column (convert categories to numbers)
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# STEP 12: Final check for missing values
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# STEP 13: View cleaned dataset
print("\nCleaned Dataset Preview:")
display(df.head())

# STEP 14: Save cleaned dataset to CSV
cleaned_file_name = 'titanic_cleaned.csv'
df.to_csv(cleaned_file_name, index=False)

print(f"\nCleaned file saved as {cleaned_file_name}")

# STEP 15: Download cleaned file
files.download(cleaned_file_name)
```
